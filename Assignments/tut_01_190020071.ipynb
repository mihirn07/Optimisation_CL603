```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
#Line Plots

x1 = np.linspace(-1,1,70)
x2 = [-1,1]
y1 = (x2[0]-x1)**4 + 8*(x1*x2[0])-x1+x2[0]+3
y2 = (x2[1]-x1)**4 + 8*(x1*x2[1])-x1+x2[1]+3
plt.plot(x1,y1,'*-g')
plt.plot(x1,y2,'|-b')
plt.grid()
plt.legend(['x2 = -1','x1 = 1'])
plt.ylabel('Test Function')
plt.xlabel('x1')
plt.title('Test Function vs x1 for a fixed x2')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12,6)
```

```python
#Surface Plot

x1 = np.linspace(-1,1,70)
x2 = np.linspace(-1,1,70)
x1,x2 = np.meshgrid(x1,x2)
func = (x2-x1)**4 + 8*(x1*x2)-x1+x2+3
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
ax.plot_surface(x1, x2, func, cmap = 'plasma')
plt.xlabel('x1')
plt.ylabel('x2')
ax.set_zlabel('Test Function')
plt.show()
```

```python
#Contour Plot

x1 = np.linspace(-1,1,70)
x2 = np.linspace(-1,1,70)
x1,x2 = np.meshgrid(x1,x2)
func = (x2-x1)**4 + 8*(x1*x2)-x1+x2+3
fig, ax = plt.subplots()
levels = np.linspace(1,9,20)
CS = ax.contour(x1, x2, func, levels)
ax.clabel(CS, inline=1, fontsize=10)
fig.set_size_inches(12,6)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot of Test Function')
plt.show()
```
