```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def input_function():

  print("Type the number of Decision Variables")
  n = int(input())

  
  print("Input for G matrix")
  G = np.zeros((n,n))
    
  for i in range(n):
    for j in range(n):
      print(f'Type ({i},{j}) element')
      G[i][j] = int(input())

  print("\nInput for d")
  d = np.zeros(n)
    
  for i in range(n):
    print(f'Type ({i+1}) element')
    d[i] = int(input())

  print("\nInput for x0")
  x0 = np.zeros(n)
    
  for i in range(n):
    print(f'Type ({i+1}) element')
    x0[i] = int(input())
    
  print("\nType number of equality constraints")
  neq = int(input())

  if neq > 0:
    
    Aeq = np.zeros((neq,n))
    beq = np.zeros((neq,1))
    
    print("\nInput for Aeq matrix")
    for i in range(neq):
      for j in range(n):
        print(f'Type ({i},{j}) element')
        Aeq[i][j] = int(input())
        
    print("\nInput for beq matrix")
    for i in range(neq):
      print(f'Type ({i+1}) element')
      beq[i] = int(input())
        
  else:
    Aeq = None
    beq = None

  print("\nType number of inequality constraints")
  nineq = int(input())

  if nineq > 0:
    Aineq = np.zeros((nineq,n))
    bineq = np.zeros((nineq,1))
    
    print("\nInput for Aineq matrix")
    
    for i in range(nineq):
      for j in range(n):
        print(f'Type ({i},{j}) element')
        Aineq[i][j] = int(input())
        
    print("\nInput for bineq matrix")
    
    for i in range(nineq):
      print(f'Type ({i+1}) element')
      bineq[i] = int(input())
        
  else:
    Aineq = None
    bineq = None

    
  return G, d, Aeq, beq, Aineq, bineq, x0, n, neq, nineq
```

```python
G, d, Aeq, beq, Aineq, bineq, x0, n, neq, nineq = input_function()
print("Printing User Input Values")
print("G ",G)
print("d ", d)
print("Aeq ", Aeq)
print("beq ", beq)
print("Aineq ", Aineq)
print("bineq ", bineq)
print("x0 ", x0)
print("n ",n)
print("neq ", neq)
print("nineq ", nineq)
```
